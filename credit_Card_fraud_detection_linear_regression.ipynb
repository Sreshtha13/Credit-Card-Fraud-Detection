```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset panda frame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset infor
credit_card_data.info()
```

```python
# checing the number of missing value in each column
credit_card_data.isnull().sum()
```

```python
credit_card_data = credit_card_data.dropna()  # Removes rows with any missing values
# OR
credit_card_data = credit_card_data.dropna(axis=1)  # Removes columns with any missing values
credit_card_data['V27'].fillna(credit_card_data['V27'].mean(), inplace=True)
credit_card_data['V28'].fillna(credit_card_data['V28'].mean(), inplace=True)
credit_card_data['Amount'].fillna(credit_card_data['Amount'].median(), inplace=True)
credit_card_data['Class'].fillna(credit_card_data['Class'].mode()[0], inplace=True)

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')  # You can also use 'median', 'most_frequent', etc.
credit_card_data['V27'] = imputer.fit_transform(credit_card_data[['V27']])
credit_card_data['V28'] = imputer.fit_transform(credit_card_data[['V28']])
credit_card_data['Amount'] = imputer.fit_transform(credit_card_data[['Amount']])
credit_card_data['Class'] = imputer.fit_transform(credit_card_data[['Class']])

```

```python
credit_card_data.isnull().sum()
```

```python
# distribution of legit and fraudelent transactions
credit_card_data['Class'].value_counts()
```

```python
# seprating the fraudelent with legit
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both the transactions
credit_card_data.groupby('Class').mean()
```

```python
# build a sample dataset of legit and fraudelent transactions
# no of fraudelent transactions >=492
legit_sample= legit.sample(n=492)

# concatenate the 2 dataset
new_dataset=pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# split data into features and targets(legit and fraud)
x=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(x)
```

```python
print(Y)
```

```python
# split data into training data and test data
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.2, stratify=Y, random_state=2)
print(x.shape, x_train.shape, x_test.shape)
```

using logistic regression


```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(x_train, Y_train)

```

```python
# accuracy on training data
X_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```
